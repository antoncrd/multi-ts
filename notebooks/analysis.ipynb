# Multidimensional Conformal Temporal Series — Interactive Analysis

This notebook provides interactive exploration of experiment results.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from src.utils.plotting import setup_style

setup_style()
%matplotlib inline
```

## Load Results

```python
# Update this path to your sweep results directory
RESULTS_DIR = Path("../results/sweep/")

records = []
for json_path in RESULTS_DIR.rglob("results.json"):
    with open(json_path) as f:
        data = json.load(f)
    record = {**data["config"], **data["calibration"], **data["metrics"]}
    records.append(record)

df = pd.DataFrame(records)
print(f"Loaded {len(df)} results")
df.head()
```

## Quick Sanity Checks

```python
if len(df) > 0:
    print(f"Mean coverage: {df['coverage'].mean():.3f}")
    print(f"Mean coverage gap: {df['coverage_gap'].mean():.4f}")
    print(f"Coverage range: [{df['coverage'].min():.3f}, {df['coverage'].max():.3f}]")
    print(f"\nDimensions: {sorted(df['d'].unique())}")
    print(f"rho_A values: {sorted(df['rho_A'].unique())}")
    print(f"k values: {sorted(df['k'].unique())}")
```
